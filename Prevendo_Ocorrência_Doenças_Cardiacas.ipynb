# Prevendo a Ocorrência de Doenças Cardíacas #

**Serviços de saúde são capazes de coletar grandes quantidades de dados sobre a saúde de seus pacientes.** 

**Um Cluster Hadoop pode ser a solução ideal para armazenar e processar esse “Big Data da Saúde”.**

### DEFINIÇÕES ###

•	**Doença cardíaca:** Condição médica crônica ou aguda que afeta um ou mais componentes do coração.

•	**Mediastino:** Lugar onde o coração está posicionado – partindo do centro do corpo humano, dois terços para a esquerda. 

•	**Artérias Coronárias:** Correm ao longo da superfície do coração e fornecem sangue rico em oxigênio ao músculo cardíaco.

•	**Pericárdio:** Teia de tecido nervoso que atravessa o coração, conduzindo os sinais neurológicos complexos que regem a contração e relaxamento.

### MACHINE LEARNING ###

• **Modelos Preditivos:** Usados para prever a ocorrência de doenças cardíacas a partir de dados coletados dos pacientes, usando algoritmos de classificação. 

### PROJETO ###

* **Estudo de caso:** Investigar todos os passos necessários na criação de um modelo preditivo doenças cardíacas. 

* **Criação de um modelo preditivo:** Utilizando variáveis preditoras que sejam capazes de prever (a partir de novos conjuntos de dados) se um paciente pode ou não desenvolver doenças do coração.

### FERRAMENTAS - ECOSSITEMA HADOOP ###

* **Hive:** – como os dados estão em formato estruturado, usaremos o Hive para armazenar os dados no HDFS e realizar consultas interativas através da linguagem HQL.

* **Pig:** – será usado para transformação e pré-processamento nos dados.

* **Mahout:** – será usado para construção do modelo preditivo.

### BASE DE DADOS ###

•	**pacientes.csv:** (Dados Coletados de um Hospital que atende Pacientes Idosos)

### ETAPAS - script.sh ###

• Etapa 1 - Carregando o dataset no Hive e visualizando os dados com SQL

• Etapa 2 - Analise Exploratória e Pré-processamento nos dados com Pig

• Etapa 3 - Transformação de Dados com o Pig

• Etapa 4 - Criação do Modelo Preditivo de Classificação

• Etapa 5 - Otimização do Modelo Preditivo de Classificação

**Etapa 1 - Carregando o dataset no Hive e visualizando os dados com SQL:**

* Carregamos os dados em uma tabela criada com o Hive. O Hive é a solução ideal para dados estruturados e permite utilizar a HQL, uma variação da linguagem SQL, que nospermite consultar os dados de forma rápida e eficiente.

**Etapa 2 - Analise Exploratória e Pré-processamento nos dados com Pig:**

* Usaremos o Pig para compreender como os dados se relacionam e realizar análises estatísticas.

**Etapa 3 - Transformação de Dados com o Pig:**

* O Pig será usado para transformar os dados de forma a facilitar o trabalho de construção do modelo preditivo.

**Etapa 4 - Criação do Modelo Preditivo de Classificação e Etapa 5 - Otimização do Modelo Preditivo de Classificação:**

* Criação do modelo preditivo com algoritmo Random Forest. 

* Inicialmente criamos um modelo e avaliamos a Confusion Matrix e na sequência otimizamos o modelo aumentando o número de árvores de decisão.
